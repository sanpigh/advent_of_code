```python
import pandas as pd
import numpy as np

with open('input') as f:
    lines_bck = f.read().splitlines()

# df = pd.read_csv('input', header=None, dtype=str)
# df
lines = lines_bck[0].split(',')
# lines = [x.replace(' -> ',',') for x in lines_bck]
# lenlines = len(lines)
# print(lenlines)3
lines = [int(x) for x in lines]
lines
```

```python
minl = min(lines)
maxl = max(lines)
print(minl, maxl)
```

```python
# Program to find most frequent
# element in a list
def most_frequent(List):
    return max(set(List), key = List.count)
 
List = [2, 1, 2, 2, 1, 3]
print(most_frequent(List))
```

```python
mf = most_frequent(lines)
```

```python
linesnp = np.asarray(lines)
(abs(linesnp-1)).sum() #- linesnp.shape[0]
```

```python
minimum = 1000000000000
for i in range(minl, maxl+1):
    dist  = abs(linesnp-i)
    fuel  = dist*(dist+1)/2
    total_fuel = fuel.sum() 
    if  total_fuel < minimum:
        minimum = total_fuel
#     print((abs(linesnp-i)).sum(), end=' ')
```
